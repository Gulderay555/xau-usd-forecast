
# Временной ряд XAU/USD — EDA и прогноз

**Цель:** Исследовать и смоделировать ежедневный обменный курс XAU/USD для прогнозирования его значения на следующий день.
Набор данных: `Daily.csv` (1978–2025, дневная детализация; несколько валют; мы сосредоточимся на **XAU/USD**).

**Результаты проекта:**
- EDA (проверка качества, тренды, сезонность, корреляции)
- Разработка признаков (лаги, скользящие средние, календарные признаки)
- Разделение обучающей и тестовой выборок с учётом времени
- Моделирование (базовое + на основе дерева)
- Метрики (MAE, RMSE, MAPE) и прогноз с фактическими данными
- Идеи для мониторинга и анализа дрейфа (не успела, были проблемы с либами)


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import evidently

from sklearn.model_selection import TimeSeriesSplit
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, root_mean_squared_error

from pathlib import Path
import warnings
warnings.filterwarnings('ignore')
```

## Загрузка и проверка данных

```python

DATA_PATH = Path("../data/Daily.csv")

# читаем данные(загружаем)
df_raw = pd.read_csv(DATA_PATH)
print(df_raw.shape)
df_raw.head()
```

## Очистка и преобразование типов

```python

df = df_raw.copy()

# Парсим дату
df['Date'] = pd.to_datetime(df['Date'], errors='coerce', infer_datetime_format=True)

# Определяем числовые столбцы, за исключением Даты
num_cols = [c for c in df.columns if c != 'Date']

# Удаляем разделитиели между тясячными долями и преобразуем в float ("45,678.90"  →  "45678.90")
for c in num_cols:
    df[c] = pd.to_numeric(df[c].astype(str).str.replace(',', ''), errors='coerce')

# Сортируем по дате и удаляем дубликаты
df = df.drop_duplicates(subset=['Date']).sort_values('Date').reset_index(drop=True)

# Оставляем данные только со столбца USD
df = df[~df['USD'].isna()].copy()

print(df.dtypes)
df.describe(include='all').T
```

## Разведочный анализ данных: анализ пропусков и основных статистических показателей

```python

#isna() NaN, mean -counts

missing_summary = df.isna().mean().sort_values(ascending=False)
missing_summary
```

```python

# Строим график XAU/USD 

plt.figure(figsize=(12,4))
plt.plot(df['Date'], df['USD'])
plt.title('XAU/USD')
plt.xlabel('Дата')
plt.ylabel('XAU/USD')
plt.show()
```

```python

# Временной анализ данных (time-series EDA). Добавляются календарные признаки(год, месяц, день недели)
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['dayofweek'] = df['Date'].dt.dayofweek

# Годовой обзор (среднее значение XAU/USD)
yearly = df.groupby('year', as_index=False)['USD'].mean()
plt.figure(figsize=(12,4))
plt.plot(yearly['year'], yearly['USD'])
plt.title('Годовое среднее значение XAU/USD')
plt.xlabel('Год')
plt.ylabel('Среднее значение XAU/USD')
plt.show()
```

## Сезонное разложение

```python

try:
    from statsmodels.tsa.seasonal import seasonal_decompose
    # даты как индекс, чтобы сделать DataFrame временным рядом.(частота- ежедневно('D'))
    df_ts = df.set_index('Date').asfreq('D')
    # Заполняем пробелы для разложения (значением из предыдущей строчки)
    df_ts['USD'] = df_ts['USD'].ffill()
    result = seasonal_decompose(df_ts['USD'], model='additive', period=365)
    result.plot()
    plt.show()
except Exception as e:
    print('Decomposition skipped:', e)
```

## Построение признаков

```python
#Этот фрагмент кода готовит датасет для обучения модели прогнозирования
work = df.copy()

# Создаем лаговые признаки (Lag features). Сдвигаем значения столбца USD вниз на lag строк, чтобы модель могла вспомнить прошлое для предсказания будущего
for lag in [1, 3, 7, 14, 30]:
    work[f'USD_lag_{lag}'] = work['USD'].shift(lag)

# Создаем скользящих средних (Moving Averages). .rolling(window=win) берёт “окно” из последних win значений, .mean() считает среднее значение внутри окна.
for win in [3, 7, 14, 30]:
    work[f'USD_ma_{win}'] = work['USD'].rolling(window=win, min_periods=1).mean()

# Очистка данных от пропусков после лагов
work = work.dropna().reset_index(drop=True)

#Разделение на признаки (X) и целевую переменную (y)
#feature_cols — список всех колонок, кроме 'USD' (цель) и 'Date'. То есть сюда входят все созданные лаги и скользящие средние.
#X — таблица с признаками (входы модели) (ФИЧИ)., y(Цели) — целевая переменная для предсказания модели (курс XAU/USD  за сегодня, например).
feature_cols = [c for c in work.columns if c not in ['USD', 'Date']]
X = work[feature_cols]
y = work['USD']

# Проверка формы данных, показывает размерности
X.shape, y.shape
```

## Разделение на обучающую и тестовую выборки (с учётом времени)

```python

# Оставляем последние 3 года для теста
last_date = work['Date'].max()
test_start = last_date - pd.Timedelta(days=365*3)

train_idx = work['Date'] < test_start
test_idx  = work['Date'] >= test_start

X_train, y_train = X[train_idx], y[train_idx]
X_test,  y_test  = X[test_idx], y[test_idx]

print('Train range:', work.loc[train_idx, 'Date'].min(), '→', work.loc[train_idx, 'Date'].max(), len(X_train))
print('Test  range:', work.loc[test_idx, 'Date'].min(),  '→', work.loc[test_idx, 'Date'].max(),  len(X_test))
```

## Моделирование: базовые модели (Linear Regression, Random Forest)

```python
#мапа для хранения обученных моделей
models = {}

# Linear Regression
lin = LinearRegression()
lin.fit(X_train, y_train)
models['LinearRegression'] = lin

# Random Forest(300 деревьев решешний)
rf = RandomForestRegressor(
    n_estimators=300,
    max_depth=None,
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train, y_train)
models['RandomForest'] = rf

# Определение метрики MAPE(Mean Absolute Percentage Error) — средняя процентная ошибка. 
#(Показывает, на сколько процентов в среднем модель ошибается относительно реального значения.)
def mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    eps = 1e-9
    return np.mean(np.abs((y_true - y_pred) / (y_true + eps))) * 100

#Проходим по всем моделям (LinearRegression, RandomForest).
#Делаем предсказания pred = model.predict(X_test).
#Считаем метрики:
#MAE (Mean Absolute Error) — средняя абсолютная ошибка;
#RMSE (Root Mean Squared Error) — корень из средней квадратичной ошибки;
#MAPE (%) — средняя процентная ошибка.

results = []
for name, model in models.items():
    pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, pred)
    rmse = root_mean_squared_error(y_test, pred)
    mp = mape(y_test, pred)
    results.append({'model': name, 'MAE': mae, 'RMSE': rmse, 'MAPE_%': mp})

#Создаем таблицу результатов, то есть results превращаем в DataFrame, сортиру по RMSE
res_df = pd.DataFrame(results).sort_values('RMSE')
res_df
```

## Сравнение прогнозов и реальных значений (на тестовом периоде)

```python

#определяем лучшую модель
best_model_name = res_df.iloc[0]['model']
best_model = models[best_model_name]

#делаем прогноз на лучшей модели
pred = best_model.predict(X_test)

plot_df = work.loc[test_idx, ['Date', 'USD']].copy()
plot_df['Pred'] = pred

plt.figure(figsize=(12,4))
plt.plot(plot_df['Date'], plot_df['Pred'], label='Предсказанное')
plt.plot(plot_df['Date'], plot_df['USD'], label='Реальное')

plt.title(f'XAU/USD — Реальное и Предсказанное значения моделью ({best_model_name})')
plt.xlabel('Дата')
plt.ylabel('XAU/USD')
plt.legend()
plt.show()
```

## Наивный прогноз на один шаг вперёд (один день)

```python

# Берём последнюю строку данных
last_row = work.iloc[-1:].copy()

# Готовим признаки для прогноза
next_date = last_row['Date'].iloc[0] + pd.Timedelta(days=1)

#Пересчитать лаговые признаки, используя последнее известное значение XAU/USD
next_features = last_row.drop(columns=['USD', 'Date']).copy()

next_pred = best_model.predict(next_features)[0]
print('Последняя дата:', last_row['Date'].iloc[0].date())
print('Слудующая дата (для прогноза):', next_date.date())
print('Предсказанное значение XAU/USD:', float(next_pred))
```
